# 1. Новый атрибут
Найдем расстояние между пунктом посадки и местом назначения. Для этого я использую модифицированную Формулу гаверсинусов, позволяющая по известным координатам (выраженным через широту и долготу) 2-ух точек найти расстояния между ними.

```python

def find_dist(x):
    #перевести координаты в радианы
    lat1 = x['pickup_latitude']*math.pi/180
    lat2 = x['dropoff_latitude']*math.pi/180
    long1 = x['pickup_longitude']*math.pi/180
    long2 = x['dropoff_longitude']*math.pi/180
    
    #косинусы и синусы широт и разницы долгот
    cl1 = math.cos(lat1)
    cl2 = math.cos(lat2)
    sl1 = math.sin(lat1)
    sl2 = math.sin(lat2)
    difL = long2 - long1
    cosDif = math.cos(difL)
    sinDif = math.sin(difL)
    
    #вычисления расстояния через модифицированную Формулу гаверсинусов
    y = ((cl2 * sinDif)**2 + (cl1 * sl2 - sl1 * cl2 * cosDif)**2)**0.5
    x = sl1 * sl2 + cl1 * cl2 * cosDif
    distRadian = math.atan(y/x)
    dist = distRadian*6372795 #умножаем на радиус земли и переводим радиусы в км
    
    
    
    '''
    sl1 = sin(lat1)
    sl2 = sin(lat2)
    delta = long2 - long1
    cdelta = cos(delta)
    sdelta = sin(delta)
    '''
    return dist
    
    
data['distantion'] = data.apply(lambda x: find_dist(x), axis=1)
data
```

Преобразуем категориальные переменные в числовые

Как видно из матрицы графиков рассеяния очень много пар можно разбить на кластеры. Я в данном работе рассмотрю зависимость признаков "prace" и "distation". Рассмотрим этот график отдельн.

```python
plt.scatter(data["price"],data["distation"])
```

Визуально можно предположить, что количество кластеров 3-4
Проанализируем теперь кол-во кластеров методом локтя (k-mean)
# Метод локтя

```python
k_range = range(1, 11)
sse = []
for k in k_range:
    km = KMeans(n_clusters = k, n_init = 10)
    km.fit(data[["price","distation"]])
    sse.append(km.inertia_)
    #metrics.silhouette_score(data, km.labels_, metric= 'euclidean', sample_size = len(data))
plt.xlabel('k')
plt.ylabel('Сумма квадратов расстояний')

plt.plot(k_range, sse)
plt.scatter(k_range, sse)
```

По графику зависимости значений суммы квадратов расстояний до центроида (весов) от количества кластеров подтверждается наше предположение. Кол-во кластерв = 3 или 4 (ближе к 3). 
Тогда будем считать что всего k = 3. Соотнесем теперь каждое наблюдение соответсвующему кластеру

```python
km = KMeans(n_clusters = 3, n_init = 10)
y_pred = km.fit_predict(data[["price","distation"]])
data["cluster"] = y_pred
data.head()
```

```python
data1 = data[data.cluster == 0]
data2 = data[data.cluster == 1]
data3 = data[data.cluster == 2]

plt.scatter(data1["price"], data1["distation"], color='green')
plt.scatter(data2["price"], data2["distation"], color='blue')
plt.scatter(data3["price"], data3["distation"], color='red')

plt.xlabel('price')
plt.ylabel('distantion')
```
